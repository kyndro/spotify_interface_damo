```python
import spotipy
import pandas as pd
```

```python
from spotipy.oauth2 import SpotifyClientCredentials
import json
```

```python
# open file with keys and set the path to your credentials JSON file
# r= read 
credentials = "example_spotify.json"
with open(credentials, "r") as keys:
    api_tokens = json.load(keys)
```

```python
api_tokens
```

```python
# read the keys and assign each to a variable
client_id = api_tokens["client_id"]
client_secret = api_tokens["client_secret"]
```

```python
#client_credentials_manager= object that hold on client id
client_credentials_manager = SpotifyClientCredentials(client_id, client_secret)
```

```python
#sp= the person relaying these messages back and forth. like a runner
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
#search
# learn about more query strings for spotify here: https://developer.spotify.com/documentation/web-api/reference/#/operations/search
track_results = sp.search(q='year:2021', type='track', limit=50)
```

```python
print(track_results)
```

```python
track_results['tracks']['items']
```

```python
tracks = track_results['tracks']['items']
```

```python
for song in tracks:
    print(song['name'])
```

```python
# convert list of dictionaries to pandas dataframe
# will not work on nested data
df = pd.DataFrame(tracks)
```
